## Visualisasi distribusi alasan prokrastinasi

### Subtask:
Analisis dan visualisasikan kolom `procrastination_reasons` untuk melihat alasan prokrastinasi yang paling umum. Karena satu baris dapat memiliki banyak alasan, kita perlu memproses kolom ini terlebih dahulu.


**Reasoning**:
The subtask requires analyzing and visualizing the distribution of procrastination reasons. The `procrastination_reasons` column contains multiple reasons in a single entry, separated by commas. To analyze the frequency of each reason, I need to split these entries and then count the occurrences of each individual reason. I will then select the top 10 reasons and visualize their frequencies using a bar plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Separate the comma-separated values in 'procrastination_reasons'
reasons = df['procrastination_reasons'].str.split(', ').explode()

# Count the frequency of each reason
reason_counts = reasons.value_counts()

# Select the top 10 reasons
top_10_reasons = reason_counts.head(10)

# Create a bar plot for the top 10 reasons
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_reasons.index, y=top_10_reasons.values, palette='viridis')
plt.title('10 Alasan Prokrastinasi Teratas', fontsize=16)
plt.xlabel('Alasan Prokrastinasi', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Analisis hubungan antara prokrastinasi dan kinerja akademik (cgpa)

### Subtask:
Analisis hubungan antara frekuensi penundaan tugas (`assignment_delay_frequency`), waktu pengumpulan tugas (`assignment_submission_timing`), atau persiapan ujian menit-menit terakhir (`last_minute_exam_preparation`) dengan CGPA. Kita perlu menangani kolom CGPA yang merupakan rentang kategori.


**Reasoning**:
Check the unique values in the 'cgpa' column to understand the categories.



```python
print("Nilai unik dalam kolom 'cgpa':")
display(df['cgpa'].unique())
```

**Reasoning**:
Map the categorical CGPA ranges to numerical values and analyze the relationship between 'assignment_delay_frequency' and the numerical CGPA using a box plot.



```python
# Create a mapping for CGPA categories to numerical values (using the midpoint of the range)
cgpa_mapping = {
    'Below 2.50': 2.25,
    '2.50 - 2.99': 2.75,
    '3.00 - 3.49': 3.25,
    '3.50 - 3.74': 3.62,
    '3.75 - 4.00': 3.87,
    None: None # Handle missing values
}

# Apply the mapping to create a new numerical CGPA column
df['cgpa_numeric'] = df['cgpa'].map(cgpa_mapping)

# Analyze the relationship between 'assignment_delay_frequency' and 'cgpa_numeric' using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='assignment_delay_frequency', y='cgpa_numeric', data=df, palette='viridis')
plt.title('Hubungan antara Frekuensi Penundaan Tugas dan CGPA', fontsize=16)
plt.xlabel('Frekuensi Penundaan Tugas', fontsize=12)
plt.ylabel('CGPA (Numerik)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Analyze the relationship between 'assignment_submission_timing' and the numerical CGPA using a box plot to further explore the link between procrastination-related behaviors and academic performance.



```python
# Analyze the relationship between 'assignment_submission_timing' and 'cgpa_numeric' using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='assignment_submission_timing', y='cgpa_numeric', data=df, palette='viridis')
plt.title('Hubungan antara Waktu Pengumpulan Tugas dan CGPA', fontsize=16)
plt.xlabel('Waktu Pengumpulan Tugas', fontsize=12)
plt.ylabel('CGPA (Numerik)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Analyze the relationship between 'last_minute_exam_preparation' and the numerical CGPA using a box plot to see if last-minute preparation is related to academic performance.



```python
# Analyze the relationship between 'last_minute_exam_preparation' and 'cgpa_numeric' using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='last_minute_exam_preparation', y='cgpa_numeric', data=df, palette='viridis')
plt.title('Hubungan antara Persiapan Ujian Menit Terakhir dan CGPA', fontsize=16)
plt.xlabel('Persiapan Ujian Menit Terakhir', fontsize=12)
plt.ylabel('CGPA (Numerik)', fontsize=12)
plt.tight_layout()
plt.show()
```

## Eksplorasi faktor lain yang mempengaruhi prokrastinasi

### Subtask:
Jelajahi hubungan antara kolom lain seperti `socio-economic_background`, `study_hours_per_week`, `hours_spent_on_mobile_non_academic`, dan `study_session_distractions` dengan kolom-kolom yang berkaitan dengan prokrastinasi.

**Reasoning**:
Explore the relationship between 'hours_spent_on_mobile_non_academic' and 'last_minute_exam_preparation' using a count plot.



```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='hours_spent_on_mobile_non_academic', hue='last_minute_exam_preparation', palette='viridis', order=['1-2 hours', '3-4 hours', 'More than 4 hours'])
plt.title('Hubungan antara Jam Penggunaan Ponsel (Non-akademik) dan Persiapan Ujian Menit Terakhir', fontsize=16)
plt.xlabel('Jam Penggunaan Ponsel (Non-akademik)', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)
plt.legend(title='Persiapan Ujian Menit Terakhir')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Explore the relationship between 'study_session_distractions' and 'assignment_delay_frequency' using a count plot.



```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='study_session_distractions', hue='assignment_delay_frequency', palette='viridis')
plt.title('Hubungan antara Gangguan Sesi Belajar dan Frekuensi Penundaan Tugas', fontsize=16)
plt.xlabel('Gangguan Sesi Belajar', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)
plt.legend(title='Frekuensi Penundaan Tugas')
plt.tight_layout()
plt.show()
```
